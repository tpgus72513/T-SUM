```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("한국전력공사_지역별 전기차 현황정보_20250228.csv")

import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
plt.rc('axes', unicode_minus=False)  # 마이너스 깨짐 방지

df["기준일"] = pd.to_datetime(df["기준일"])
df = df.sort_values("기준일")

plt.plot(df["기준일"], df["세종"], marker="o")
plt.title("세종 전기차 등록대수 증가 추세 (2024-08 ~ 2025-02)")
plt.xlabel("기준일")
plt.ylabel("전기차 등록 대수")
plt.grid(True)
plt.tight_layout()
plt.show()
```
